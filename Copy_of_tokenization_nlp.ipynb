```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

text = "Hello there! How are you doing today? I hope you're learning NLP."
doc = nlp(text)

print("Word Tokenization:")
for token in doc:
    print(token.text)
```
