```python
# Simple manual morphology demo
words = ["unhappiness", "redoing", "cats", "player"]

def simple_morph(word):
    if word.endswith("ing"):
        return f"{word} -> root: {word[:-3]}, suffix: ing"
    elif word.endswith("s"):
        return f"{word} -> root: {word[:-1]}, suffix: s (plural)"
    elif word.startswith("un"):
        return f"{word} -> prefix: un, root: {word[2:]}"
    elif word.startswith("re"):
        return f"{word} -> prefix: re, root: {word[2:]}"
    else:
        return f"{word} -> no clear affix"

for w in words:
    print(simple_morph(w))
```
