```python
from collections import Counter

# Training corpus
corpus = "I love NLP and I love Machine Learning".split()

# Build bigram counts
bigrams = [(corpus[i], corpus[i+1]) for i in range(len(corpus)-1)]
bigram_counts = Counter(bigrams)
unigram_counts = Counter(corpus)
vocab = set(corpus)
V = len(vocab)

def bigram_prob(w1, w2):
    return (bigram_counts[(w1, w2)] + 1) / (unigram_counts[w1] + V)

# Test probability
print("P(NLP | love) =", bigram_prob("love", "NLP"))
print("P(Learning | NLP) =", bigram_prob("NLP", "Learning"))
print("P(something | NLP) =", bigram_prob("NLP", "something"))  # unseen word
```
