```python
import spacy

# Load English model
nlp = spacy.load("en_core_web_sm")

text = "The striped bats are hanging on their feet and eating best fruits"
doc = nlp(text)

print("Lemmatization using spaCy:")
for token in doc:
    print(token.text, "->", token.lemma_)
```
